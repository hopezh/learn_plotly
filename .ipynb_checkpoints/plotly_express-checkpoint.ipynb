```python
import plotly.express as px

df = px.data.gapminder()

df.head()
```

```python
df2007 = df.query('year == 2007')
df2007.head()
```

```python
px.histogram(df, x='lifeExp')
```

```python
px.histogram(df2007, x='lifeExp')
```

```python
px.histogram(df2007, x='lifeExp', color='continent')
```

```python
px.histogram(
    df2007, 
    x='lifeExp', 
    color='continent', 
    marginal='rug', 
    hover_name='country', 
    hover_data=df2007.columns
)
```

```python
px.histogram(
    df2007, 
    x='lifeExp', 
    color='continent', 
    marginal='rug', 
    hover_name='country', 
    hover_data=df2007.columns,
    y='pop', 
    histfunc='sum'
)
```

```python
px.histogram(
    df2007, 
    x='continent', 
    hover_name='country', 
    hover_data=df2007.columns,
    y='pop', 
    histfunc='sum'
)
```

```python
px.bar(
    df2007, 
    x='continent', 
    color='lifeExp', 
    hover_name='country', 
    hover_data=df2007.columns,
    y='pop', 
)
```

```python
px.sunburst(
    df2007, 
    path=['continent', 'country'],
    color='lifeExp', 
    hover_data=df2007.columns,
    values='pop'
)
```

```python
px.choropleth(
    df2007, 
    locations='iso_alpha', 
    color='lifeExp', 
    hover_data=df2007.columns
)
```

```python
px.scatter(
    df2007, 
    x='gdpPercap', 
    y='lifeExp', 
    log_x=True,
    color='continent', 
    size='pop', 
    size_max=45, 
    facet_col='continent', 
    facet_col_wrap=3, 
    hover_data=df2007.columns, 
    hover_name='country'
)
```

```python
px.scatter(
    df, 
    x='gdpPercap', 
    y='lifeExp', 
    log_x=True,
    color='continent', 
    size='pop', 
    size_max=45, 
    animation_frame='year', 
    animation_group='country', 
    range_y=[20, 100], 
    hover_data=df2007.columns, 
    hover_name='country'
)
```
